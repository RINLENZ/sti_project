```python
# notebook_audio.ipynb
# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 1 — Imports
# ─────────────────────────────────────────────────────────────────────────────
import os, random
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from pathlib import Path
import seaborn as sns

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import classification_report, confusion_matrix

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device :", DEVICE)

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 2 — Configuration
# ─────────────────────────────────────────────────────────────────────────────
DATA_DIR = Path("../backend/data/audio")

COMMANDES = ["aide", "oui", "non", "repeter",
             "incompris", "lentement", "bruit_silence"]
CMD2IDX   = {c: i for i, c in enumerate(COMMANDES)}

for cmd in COMMANDES:
    n = len(list((DATA_DIR / cmd).glob("*.wav")))
    print(f"  {cmd:20s}: {n} clips")

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 3 — Extraction des features MFCC
#
# POURQUOI les MFCC ?
#   - Mel-Frequency Cepstral Coefficients = représentation compacte du son
#   - Mimétisent la perception humaine (échelle mel = non-linéaire)
#   - 40 coefficients par frame → résumé de la fréquence à chaque instant
#   - Résultat : image 2D (40 × temps) → CNN peut l'analyser comme une image
#
# PARAMÈTRES :
#   sr=16000  : 16 kHz standard pour la voix
#   n_mfcc=40 : 40 coefficients (suffisant pour la voix)
#   n_fft=512 : taille de la fenêtre FFT (32 ms à 16kHz)
#   hop=160   : pas de 10 ms → bonne résolution temporelle
#   MAX_LEN   : on fixe tous les spectrogrammes à la même longueur
#               (pad avec 0 si trop court, coupe si trop long)
# ─────────────────────────────────────────────────────────────────────────────
SR      = 16_000
N_MFCC  = 40
N_FFT   = 512
HOP     = 160
MAX_LEN = 100    # ~1 seconde à hop=160 → 100 frames temporelles

def extract_mfcc(path: Path) -> np.ndarray:
    """
    Charge un fichier WAV et retourne le spectrogramme MFCC normalisé.
    Shape de sortie : (1, N_MFCC, MAX_LEN) = (canal, hauteur, largeur)
    → compatible avec un CNN 2D
    """
    y, sr = librosa.load(path, sr=SR, mono=True)

    # Normalise l'amplitude (certains micros sont plus forts que d'autres)
    y = librosa.util.normalize(y)

    # Extrait les MFCC
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC,
                                  n_fft=N_FFT, hop_length=HOP)

    # Pad ou coupe pour uniformiser la longueur
    if mfcc.shape[1] < MAX_LEN:
        mfcc = np.pad(mfcc, ((0, 0), (0, MAX_LEN - mfcc.shape[1])))
    else:
        mfcc = mfcc[:, :MAX_LEN]

    # Normalise par mean/std → le CNN converge plus vite
    mfcc = (mfcc - mfcc.mean()) / (mfcc.std() + 1e-8)

    return mfcc[np.newaxis, :, :]   # ajoute dimension canal

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 4 — Visualisation d'un MFCC (pour comprendre ce qu'on donne au CNN)
# ─────────────────────────────────────────────────────────────────────────────
exemple_path = next((DATA_DIR / "aide").glob("*.wav"))
mfcc_ex = extract_mfcc(exemple_path)[0]   # retire dim canal pour affichage

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc_ex, x_axis="time", sr=SR, hop_length=HOP)
plt.colorbar(); plt.title("MFCC — exemple 'aide'"); plt.tight_layout(); plt.show()

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 5 — Dataset
# ─────────────────────────────────────────────────────────────────────────────
class AudioDataset(Dataset):
    def __init__(self, root, commandes):
        self.samples = []
        for cmd in commandes:
            for wav in (root / cmd).glob("*.wav"):
                self.samples.append((wav, CMD2IDX[cmd]))
        random.shuffle(self.samples)

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        mfcc = extract_mfcc(path)
        return torch.tensor(mfcc, dtype=torch.float32), label

full_ds = AudioDataset(DATA_DIR, COMMANDES)
n_val   = int(len(full_ds) * 0.2)
n_train = len(full_ds) - n_val
train_ds, val_ds = torch.utils.data.random_split(
    full_ds, [n_train, n_val],
    generator=torch.Generator().manual_seed(42)
)

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
val_loader   = DataLoader(val_ds,   batch_size=16, shuffle=False)
print(f"Train : {len(train_ds)} | Val : {len(val_ds)}")

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 6 — Architecture CNN légère
#
# Input : (batch, 1, 40, 100) — 1 canal, 40 MFCC, 100 frames temporelles
#
# BLOC CONV :
#   Conv2d → BatchNorm → ReLU → MaxPool
#   - Conv2d capture les patterns locaux (phonèmes, voyelles)
#   - BatchNorm accélère la convergence
#   - MaxPool réduit la dimension spatiale (résumé)
#
# CLASSIFICATION :
#   flatten → Dense → Dropout → Dense(7) → sortie
# ─────────────────────────────────────────────────────────────────────────────
class AudioCNN(nn.Module):
    def __init__(self, n_classes):
        super().__init__()
        self.features = nn.Sequential(
            # Bloc 1 : détecte les patterns basses fréquences
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),          # 40×100 → 20×50

            # Bloc 2 : patterns plus complexes (combinaisons de fréquences)
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),          # 20×50 → 10×25

            # Bloc 3 : représentation abstraite du son
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4)),   # taille fixe quelle que soit l'entrée
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 4 * 4, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, n_classes),
        )

    def forward(self, x):
        return self.classifier(self.features(x))

model = AudioCNN(len(COMMANDES)).to(DEVICE)
total_params = sum(p.numel() for p in model.parameters())
print(f"Paramètres : {total_params:,}")

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 7 — Entraînement
# ─────────────────────────────────────────────────────────────────────────────
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)

def run_epoch(loader, train=True):
    model.train(train)
    total_loss, correct, total = 0, 0, 0
    with torch.set_grad_enabled(train):
        for x, y in loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            logits = model(x)
            loss   = criterion(logits, y)
            if train:
                optimizer.zero_grad(); loss.backward(); optimizer.step()
            total_loss += loss.item() * len(y)
            correct    += (logits.argmax(1) == y).sum().item()
            total      += len(y)
    return total_loss / total, correct / total

EPOCHS = 30
for epoch in range(1, EPOCHS + 1):
    tl, ta = run_epoch(train_loader, True)
    vl, va = run_epoch(val_loader,   False)
    scheduler.step()
    if epoch % 5 == 0 or epoch == 1:
        print(f"Epoch {epoch:2d} | train {ta:.2%}  val {va:.2%}")

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 8 — Évaluation
# ─────────────────────────────────────────────────────────────────────────────
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for x, y in val_loader:
        preds = model(x.to(DEVICE)).argmax(1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(y.numpy())

print(classification_report(all_labels, all_preds, target_names=COMMANDES))

cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=COMMANDES, yticklabels=COMMANDES, cmap="Greens")
plt.title("Matrice de confusion — Commandes audio")
plt.ylabel("Réel"); plt.xlabel("Prédit")
plt.tight_layout(); plt.show()

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 9 — Test sur un fichier réel
# ─────────────────────────────────────────────────────────────────────────────
def predict_audio(wav_path: str) -> str:
    mfcc = extract_mfcc(Path(wav_path))
    x    = torch.tensor(mfcc, dtype=torch.float32).unsqueeze(0).to(DEVICE)
    with torch.no_grad():
        probs = torch.softmax(model(x), dim=1)[0].cpu().numpy()
    for cmd, p in sorted(zip(COMMANDES, probs), key=lambda t: -t[1]):
        print(f"  {cmd:15s}: {p:.1%}")
    return COMMANDES[probs.argmax()]

# test sur un fichier aléatoire
test_file = next((DATA_DIR / "oui").glob("*.wav"))
print(f"\nTest sur : {test_file.name}")
pred = predict_audio(test_file)
print(f"→ Prédit : {pred}")

# ─────────────────────────────────────────────────────────────────────────────
# CELLULE 10 — Sauvegarde
# ─────────────────────────────────────────────────────────────────────────────
torch.save({
    "model_state": model.state_dict(),
    "commandes":   COMMANDES,
    "n_mfcc":      N_MFCC,
    "max_len":     MAX_LEN,
    "sr":          SR,
    "hop":         HOP,
}, "model_audio.pth")
print("Modèle sauvegardé → model_audio.pth")
```
